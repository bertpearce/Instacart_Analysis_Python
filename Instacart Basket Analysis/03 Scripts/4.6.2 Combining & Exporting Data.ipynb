## 04. Merge data

```python
# Since our key questions require data that is unique to both datasets, we'll need to use outer join to retain all data.

# Preview the merge of dataframe df_ords_prods_combined and df_prods_prepared with an indicator column

pd.merge(df_ords_prods_combined, df_prods_prepared, on = ['product_id'], how = 'outer', indicator='2_merge')
```

```python
# Review the value frequencies of the merge flag for verification

pd.merge(df_ords_prods_combined, df_prods_prepared, on = ['product_id'], how = 'outer', indicator='2_merge')['2_merge'].value_counts()
```

```python
# Create new dataframe df_ords_prods_merge by merging dataframes DF1 and DF2 based on the shared columns COLUMN1, COLUMN2 using the default inner join

df_ords_prods_merge = df_ords_prods_combined.merge(df_prods_prepared, on = ['product_id'], how = 'outer')
```

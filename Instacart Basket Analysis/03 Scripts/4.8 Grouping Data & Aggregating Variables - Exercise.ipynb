### Exercise Step 4

```python
# Group and transform user_id

df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')

print(f"New variable created.")
```

```python
# Review DataFrame

df_ords_prods_merge.head()
```

```python
# Populate the column using a loc() function.

df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'

print(f"Column populated with loc() function")
```

```python
# Review DataFrame

df_ords_prods_merge.head()
```

```python
# Verify the value counts

df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

### Exercise Step 5

```python
# Group and aggregate the variable

df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

### Exercise Step 6

```python
# Group and transform prices variable

df_ords_prods_merge['avg_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')

print(f"New variable created.")
```

```python
# Review DataFrame

df_ords_prods_merge.head()
```

```python
# Populate the column using a loc() function.

df_ords_prods_merge.loc[df_ords_prods_merge['avg_price'] < 10, 'spending_type'] = 'Low spender'
df_ords_prods_merge.loc[df_ords_prods_merge['avg_price'] >= 10, 'spending_type'] = 'High spender'

print(f"Column populated with loc() function")
```

```python
# Review DataFrame

df_ords_prods_merge.head()
```

```python
# Verify the value counts

df_ords_prods_merge['spending_type'].value_counts(dropna = False)
```

### Exercise Step 7

```python
# Group and transform user_id

df_ords_prods_merge['order_frequency'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')

print(f"New variable created.")
```

```python
# Review DataFrame

df_ords_prods_merge.head()
```

```python
# Populate the column using a loc() function.

df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] > 20, 'customer_frequency'] = 'Non-frequent customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['order_frequency'] <= 20) & (df_ords_prods_merge['order_frequency'] > 10), 'customer_frequency'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['order_frequency'] <= 10, 'customer_frequency'] = 'Frequent customer'

print(f"Column populated with loc() function")
```

```python
# Review DataFrame

df_ords_prods_merge.head()
```

```python
# Verify the value counts

df_ords_prods_merge['customer_frequency'].value_counts(dropna = False)
```
